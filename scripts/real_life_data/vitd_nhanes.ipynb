```python
import pandas as pd
import pyreadstat
from measures import *

#nhanes 2015-2016

data_vitd, meta_vitd = pyreadstat.read_xport("../../data/vitd/nhanes/VID_I.XPT")
data_alcohol, meta_alcohol = pyreadstat.read_xport("../../data/vitd/nhanes/ALQ_I.XPT")
data_demographics, meta_demographics = pyreadstat.read_xport("../../data/vitd/nhanes/DEMO_I.XPT")
data_physical_activity, meta_physical_activity = pyreadstat.read_xport("../../data/vitd/nhanes/PAQ_I.XPT")
data_smoking, meta_smoking = pyreadstat.read_xport("../../data/vitd/nhanes/SMQ_I.XPT")
data_body_measures, meta_body_measures = pyreadstat.read_xport("../../data/vitd/nhanes/BMX_I.XPT")
data_depression, meta_depression = pyreadstat.read_xport("../../data/vitd/nhanes/DPQ_I.XPT")
data_income, meta_income = pyreadstat.read_xport("../../data/vitd/nhanes/INQ_I.XPT")
```

```python
merged_dataset = pd.merge(data_vitd, data_alcohol, on='SEQN', how='inner')
merged_dataset = pd.merge(merged_dataset, data_demographics, on='SEQN', how='inner')
merged_dataset = pd.merge(merged_dataset, data_smoking, on='SEQN', how='inner')
merged_dataset = pd.merge(merged_dataset, data_body_measures, on='SEQN', how='inner')
merged_dataset = pd.merge(merged_dataset, data_depression, on='SEQN', how='inner')
merged_dataset = pd.merge(merged_dataset, data_income, on='SEQN', how='inner')
```

```python
dataset = merged_dataset.loc[:, [bmi, gender, age, vitamin_d, poverty_index, feel_depressed, smoke_now]]
```

```python
dataset = dataset.dropna()
```

```python
dataset = dataset[dataset[age] != 80]
dataset = dataset[dataset[vitamin_d] < 150]
dataset = dataset[dataset[feel_depressed] < 4]
#dataset = dataset[dataset[poverty_index] < 5]
dataset
```

```python
dataset.to_csv('../../data/vitd/nhanes.csv', index=False)
```
